# R 语言简介

## 基本命令

```r
x <- c(1, 3, 2, 5)
print(x)
```

```r
x = c(1, 6, 2)
print(x)
```

```r
y <- c(1, 4, 3)
```

```r
print(length(x))
```

```r
print(length(y))
```

```r
print(x + y)
```

`ls()` 函数查看所有的对象列表。

`rm()` 函数可以删除不需要的对象。

```r
print(ls())
```

```r
rm(x, y)
```

```r
print(ls())
```

```r
rm(list=ls())
```

`matrix()` 函数创建矩阵

```r
x = matrix(
    data=c(1, 2, 3, 4),
    nrow=2,
    ncol=2
)
print(x)
```

```r
x = matrix(
    c(1, 2, 3, 4),
    2,
    2
)
print(x)
```

创建矩阵时默认先填列再填行，使用 `byrow=TRUE` 表示先填行再填列

```r
print(matrix(
    c(1, 2, 3, 4),
    2,
    2,
    byrow=TRUE
))
```

```r
print(sqrt(x))
```

```r
print(x^2)
```

`rnorm()` 函数产生随机正态分布的向量

`cor()` 计算变量之间的相关系数

```r
x = rnorm(50)
y = x + rnorm(50, mean=50, sd=.1)
print(cor(x, y))
```

`set.seed()` 设置随机数种子

```r
set.seed(1303)
print(rnorm(50))
```

`mean()` 计算均值

`var()` 计算方差

`sd()` 计算标准差

```r
set.seed(3)
y = rnorm(100)
print(mean(y))
```

```r
print(var(y))
```

```r
print(sqrt(var(y)))
```

```r
print(sd(y))
```

## 图形

`plot()` 函数

```r
x = rnorm(100)
y = rnorm(100)
plot(x, y)
```

```r
plot(
    x, y,
    xlab="this is the x-axis",
    ylab="this is the y-axis",
    main="Plot of X vs Y",
)
```

使用 `pdf()` 或 `jpeg()` 等函数保存图片

```r
pdf("Figure.pdf")
plot(x, y, col="green")
dev.off()
```

`seq()` 用于生成序列

```r
x = seq(1, 10)
print(x)
```

```r
x = 1:10
print(x)
```

```r
x = seq(-pi, pi, length=50)
```

`contour` 函数用于绘制等值线图

```r
y = x
f = outer(x, y, function(x,y) cos(y)/(1 + x^2))
```

```r
contour(x, y, f)
contour(
    x, y, f, 
    nlevels=45,
    add=T,
)
```

```r
fa = (f - t(f)) / 2
```

```r
contour(
    x, y, fa,
    nlevels=15
)
```

`image()` 函数绘制 heatmap

```r
image(x, y, fa)
```

`persp()` 函数绘制三维图

```r
persp(x, y, fa)
```

```r
persp(x, y, fa, theta=30)
```

```r
persp(x, y, fa, theta=30, phi=20)
```

```r
persp(x, y, fa, theta=30, phi=70)
```

```r
persp(x, y, fa, theta=30, phi=40)
```

## 其他的图形和数值汇总

使用 `plot()` 绘制定量变量的散点图

```r
plot(Auto$cylinders, Auto$mpg)
```

```r
attach(Auto)
```

```r
plot(cylinders, mpg)
```

`as.factor` 将定量变量转换成定性变量

```r
cylinders = as.factor(cylinders)
```

```r
plot(cylinders, mpg)
```

```r
plot(cylinders, mpg, col="red")
```

```r
plot(
    cylinders, mpg, 
    col="red", 
    varwidth=T
)
```

```r
plot(
    cylinders, mpg, 
    col="red", 
    varwidth=T, 
    horizontal=T
)
```

```r
plot(
    cylinders, mpg, 
    col="red", 
    varwidth=T, 
    xlab="cylinders",
    ylab="MPG"
)
```

`hist()` 函数绘制直方图

```r
hist(mpg)
```

```r
hist(mpg, col=2)
```

```r
hist(mpg, col=2, breaks=15)
```

`pairs()` 函数建立数据集中每一对变量的散点图矩阵

```r
# 会抛出错误
# pairs(Auto)
```

```r
pairs(~ mpg + displacement + horsepower + weight + acceleration, Auto)
```

`identify()` 函数用于交互

```r
plot(horsepower, mpg)
identify(horsepower, mpg, name)
```

`summary()` 显示数据汇总信息

```r
summary(Auto)
```

```r
summary(mpg)
```
