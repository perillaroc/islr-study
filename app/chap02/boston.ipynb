## 查看数据

数据有 506 行，14 列

```r
Boston
```

```r
?Boston
```

每列的含义：

- `crim`：城镇人均犯罪率
- `zn`：占地 25,000 平方英尺以上的住宅用地比例
- `indus`：每个城镇非零售业务英亩面积
- `chas`：查尔斯河虚拟变量（如果在河流边，则为 1；否则为 0）
- `nox`：氮氧化物浓度（百万分之几）
- `rm`：每个住宅的平均房间数
- `age`：1940年之前建造的自有住房的比例
- `dis`：到五个波士顿就业中心的距离的加权平均值
- `rad`：径向公路的可及性指数
- `tax`：每 10,000 美元的全值财产税率
- `ptratio`：各镇师生比例
- `black`：1000(Bk - 0.63)^2，其中 Bk 是按城镇划分的黑人比例
- `lstat`：较低地位的人口（百分比）。
- `medv`: 拥有价值 1000 美元住房的中位数。
