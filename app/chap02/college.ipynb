# College 数据集

## 加载

载入数据集

```r
college <- read.csv(
    "../../data/College.csv",
)
head(college)
```

## 处理

使用 `fix()` 函数观察数据

注：或者使用 `head()` 函数打印头几行数据

```r
fix(college)
```

第一列数据是大学的名字

使用 `rownames()` 函数为每行设置名称

```r
rownames(college) <- college[,1]
head(college)
```

将不需要分析的第 1 列剔除

```r
college <- college[,-1]
head(college)
```

使用 `fix()` 查看数据，可以发现每列的名称被添加到显示表格中的第一列 row.names

```r
fix(college)
```

## 分析

`summary()` 函数给出汇总统计信息

```r
summary(college)
```

使用 `paris()` 函数绘制前 10 列数据的散点图矩阵

注：第一列为文本，无法绘制散点图

```r
pairs(college[, 2:11])
```

使用 `plot()` 函数绘制 Outstate 对 Private 变量的沿边箱线图

需要首先将 Private 列转换成定性变量

```r
attach(college)
Private <- as.factor(Private)
plot(Private, Outstate)
```

创建新的定性变量 Elite，合并 (binning) Top10perc 变量。

根据是否有超过 50% 以上的学生来自排名在前 10% 顶尖高中的情况将大学分成两组

```r
Elite <- rep("No", nrow(college))
Elite[college$Top10perc > 50] <- "Yes"
Elite <- as.factor(Elite)
print(head(Elite))
```

```r
college <- data.frame(college, Elite)
head(college)
```

使用 `summary()` 了解有多少个精英大学

```r
summary(Elite)
```

使用 `plot()` 产生 Outstate 对 Elite 的沿边箱线图

```r
plot(
    Elite, Outstate,
    xlab="Elite",
    ylab="Outstate"
)
```
